```python
import numpy as np
import pandas as pd
import ast
```

```python
movies3= pd.read_csv('tmdb_5000_movies.csv')
credits= pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies3.head(5)
```

```python
credits.head(1)
```

```python
movies3=movies3.merge(credits, on='title')
movies3.head(1)
```

```python
movies3=movies3[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies3.head(1)
```

```python
movies3.isnull().sum()
```

```python
movies3.dropna(inplace=True)
```

```python
movies3.isnull().sum()
```

```python
movies3.iloc[0].genres
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):    #converting string into list
        L.append(i['name'])
    return L
```

```python
movies3['genres']=movies3['genres'].apply(convert)
movies3.head(1)
```

```python
movies3['keywords'] = movies3['keywords'].apply(convert)
movies3.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        #converting string into list
        if(counter<3):
            L.append(i['name'])
            counter+=1
    return L
```

```python
movies3['cast']=movies3['cast'].apply(convert3)
```

```python
movies3.head(1)
```

```python
movies3['cast'] = movies3['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj): 
        #converting string into list
        if i['job']== 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies3['crew']=movies3['crew'].apply(fetch_director)
```

```python
movies3.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies3['cast'] = movies3['cast'].apply(collapse)
movies3['crew'] = movies3['crew'].apply(collapse)
movies3['genres'] = movies3['genres'].apply(collapse)
movies3['keywords'] = movies3['keywords'].apply(collapse)
```

```python
movies3['overview'] = movies3['overview'].apply(lambda x:x.split())
```

```python
movies3.head()
```

```python
movies3['tags']= movies3['overview']+movies3['genres']+ movies3['keywords']+movies3['cast']+movies3['crew']
```

```python
movies3.head()
```

```python
new_mov=movies3[['movie_id', 'title', 'tags']]
```

```python
new_mov
```

```python
new_mov['tags']=new_mov['tags'].apply(lambda x:" ".join(x))
```

```python
new_mov.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_mov['tags']=new_mov['tags'].apply(stem)
```

```python
new_mov['tags']=new_mov['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_mov['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])), reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index= new_mov[new_mov['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_mov.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
